### Soru 7
Aşağıdaki iki DataFrame'i **"id"** sütununa göre birleştirin (merge) ve sonucu yazdırın:

```python
df1 = pd.DataFrame({
    "id": [1, 2, 3, 4],
    "ürün": ["Laptop", "Telefon", "Tablet", "Kulaklık"]
})

df2 = pd.DataFrame({
    "id": [1, 2, 3, 4],
    "fiyat": [15000, 10000, 5000, 500],
    "stok": [30, 50, 25, 100]
})
```

Birleştirme sonrasında **fiyatı 5000'den büyük** olan ürünleri filtreleyin.

````python
# Cevabınızı buraya yazın
df1 = pd.DataFrame({
    "id": [1, 2, 3, 4],
    "ürün": ["Laptop", "Telefon", "Tablet", "Kulaklık"]
})

df2 = pd.DataFrame({
    "id": [1, 2, 3, 4],
    "fiyat": [15000, 10000, 5000, 500],
    "stok": [30, 50, 25, 100]
})

result = pd.merge(df1,df2,on="id",how="inner")
print(result[result['fiyat']>5000])
````

## Bölüm 3: Matplotlib

### Soru 8
Aşağıdaki verileri kullanarak bir **çizgi grafiği (line plot)** çizin:

```python
aylar = ["Oca", "Şub", "Mar", "Nis", "May", "Haz"]
satislar = [120, 135, 148, 160, 175, 190]
```

Grafik gereksinimleri:
- Başlık: "Aylık Satış Trendi"
- X ekseni etiketi: "Aylar"
- Y ekseni etiketi: "Satış Adedi"
- Çizgi rengi kırmızı, işaretçi (marker) yuvarlak olsun

````python
# Cevabınızı buraya yazın
aylar = ["Oca", "Şub", "Mar", "Nis", "May", "Haz"]
satislar = [120, 135, 148, 160, 175, 190]
plt.plot(aylar,satislar,color="red",marker="o")
plt.title("Aylık Satış Trendi")
plt.xlabel("Aylar")
plt.ylabel("Satış Adedi")
````

### Soru 9
Aşağıdaki verileri kullanarak bir **bar (çubuk) grafiği** çizin:

```python
diller = ["Python", "JavaScript", "Java", "C++", "Go"]
kullanici = [35, 25, 20, 12, 8]
```

Grafik gereksinimleri:
- Başlık: "Programlama Dilleri Popülerliği"
- Her çubuk farklı renkte olsun
- Figür boyutu 10×6 olsun

````python
# Cevabınızı buraya yazın
plt.figure(figsize=(10,6))
diller = ["Python", "JavaScript", "Java", "C++", "Go"]
kullanici = [35, 25, 20, 12, 8]
renkler = ['red','blue','green','purple','orange']
plt.bar(diller,kullanici,color=renkler)
plt.title("Programlama Dilleri Popülerliği")
plt.xlabel("Diller")
plt.ylabel("Kullanici Sayisi")
plt.show()
````

## Bölüm 4: Seaborn

### Soru 10
Seaborn'un yerleşik **"tips"** veri setini yükleyin (`sns.load_dataset("tips")`).

1. İlk 5 satırı gösterin.
2. `total_bill` sütunu için bir **histogram (histplot)** çizin.
3. Grafik başlığını "Toplam Hesap Dağılımı" olarak ayarlayın ve KDE eğrisini ekleyin.

````python
# Cevabınızı buraya yazın
tips = sns.load_dataset("tips")
print(tips.head())
sns.histplot(tips["total_bill"],kde=True)
plt.title("Toplam Hesap Dağılımı")
plt.show()
````

### Soru 11
Tips veri setini kullanarak **`total_bill`** ve **`tip`** arasındaki ilişkiyi gösteren bir **scatter plot (saçılım grafiği)** çizin (`sns.scatterplot`).

Grafik gereksinimleri:
- Noktaları **`smoker`** (sigara içen/içmeyen) kategorisine göre renklendirin (hue).
- Başlık: "Hesap vs Bahşiş İlişkisi"
- Figür boyutu 10×6 olsun

````python
# Cevabınızı buraya yazın
plt.figure(figsize=(10,6))
sns.scatterplot(data=tips,x="total_bill",y="tip",hue="smoker")
plt.title("Hesap vs Bahşiş İlişkisi")
````
