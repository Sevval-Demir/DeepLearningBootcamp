```python
def salary_category(salary):
    if salary < 50000:
        return "Low"
    elif 50000 <= salary < 80000:
        return "Medium"
    else:
        return "High"
```

```python
df["Salary"].apply(salary_category)
```

```python
df["Salary_Category"] = df["Salary"].apply(salary_category)
```

```python
df
```

```python
def performance_score_func(experience):
    if experience > 10:
        return 1
    else:
        return 0
```

```python
df["New_Performance_Score"] = df["Experience"].apply(performance_score_func)
```

```python
df
```

```python
df["New_Score"] = df["Performance_Score"] + df["New_Performance_Score"]
```

```python
df["Formatted_Name"]=df["Name"].apply(lambda x : x.replace("_"," "))
```
